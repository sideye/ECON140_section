```python
import pandas as pd
import statsmodels.api as sm
```

```python
# Read in data set
cars = pd.read_csv('cars.csv')
cars.head()
```

```python
# pt a
price = cars['price']
weight = sm.add_constant(cars['weight'])
model_a = sm.OLS(price, weight)
results_a = model_a.fit(cov_type = 'HC1')
results_a.summary()
```

```python
# pt b
X_b = sm.add_constant(cars[['horsepower', 'citympg', 'cylinders', 'wheelbase', 'hybrid', 'weight']])
model_b = sm.OLS(price, X_b)
results_b = model_b.fit(cov_type = 'HC1')
results_b.summary()
```

```python
# pt a
price = cars['price']
X_c = sm.add_constant(cars[['domestic', 'european', 'asian']])
model_c = sm.OLS(price, X_c)
results_c = model_c.fit(cov_type = 'HC1')
results_c.summary()
```
